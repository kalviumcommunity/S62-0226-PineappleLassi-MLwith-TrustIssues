```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("users.csv")

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)
```

```python
sns.countplot(x='role', data = df)
plt.title('Distribution of User Roles')
plt.show()
```

```python
sns.countplot(x='department', data = df[df['role'] == 'user'])
plt.title('Distribution of User Departments')
plt.show()
```

```python
sns.histplot(df["tenure_months"], bins=10, kde=True)
plt.title("Tenure Distribution (Months)")
plt.show()
```

```python
sns.countplot(x="typical_login_hour", data=df)
plt.title("Typical Login Hour Distribution")
plt.show()
```

```python
sns.scatterplot(
    x="role",
    y="login_hour_std",
    hue="role",
    data=df
)
plt.title("Typical Login Hour vs Login Hour Standard Deviation")
plt.show()
```
